# Modelo de Teste
Simulação básica do comportamento esperado do chatbot GoodCharge AI gerenciando os carregadores elétricos em condomínios, possibilitando que os usuários façam as perguntas que quiserem.

# Objetivo
Demonstrar perguntas esperadas dos usuários e as respostas ideais do chatbot.

# Perguntas e respostas esperadas
-- Perguntas

- Quanto consumi este mês?
- Tem algum carregador livre?
- Quero reservar um carregador
- Tem algum carregador sem funcionar?
- Quanto tenho que pagar este mês?

-- Respostas

- Você consumiu 120 kWh neste mês.
- O carregador 2 não tem nenhuma reserva até as 18h.
- Reserva realizada para hoje às 20h.
- O carregador 4 encontra-se offline.
-  O valor total da cobrança é R$ 154,20.

```python
!pip install openai
from openai import OpenAI
from google.colab import userdata

client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))
```

```python
user_pergunta = input("Digite sua pergunta para o chatbot: ")

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "system",
            "content": "Você é um assistente técnico especializado em carregadores elétricos de condomínios. Responde de forma clara e curta."
        },{
            "role": "user",
            "content": "Quanto consumi este mês?"
        },{
            "role": "assistant",
            "content": "Você consumiu 120 kWh neste mês."
        },{
            "role": "user",
            "content": "Tem carregador livre?"
        },{
            "role": "assistant",
            "content": "O carregador 2 não tem nenhuma reserva até as 18h."
        },{
            "role": "user",
            "content": "Quero reservar um garregador."
        },{
            "role": "assistant",
            "content": "Reserva realizada para hoje as 20h."
        },{
            "role": "user",
            "content": "Tem algum carregador sem funcionar?"
        },{
            "role": "assistant",
            "content": "O carregador 4 encontra-se offline."
        },{
            "role": "user",
            "content": "Quanto tenho que pagar esse mês?"
        },{
            "role": "assistant",
            "content": "O valor total da cobrança é R$ 154,20."
        },{
            "role": "user",
            "content": user_pergunta
        }
    ],

    max_tokens=200
)

print(response.choices[0].message.content)
```
